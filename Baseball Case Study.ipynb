# 1. ZSCORE Method

```python
df.shape
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
threshold=3
np.where(z>3)
```

```python
df_new=df[(z<3).all(axis=1)]
```

###### So We have a data loss of 3.3333333333333335% by using ZSCORE method

# 2. IQR (Inter Quantile Range) method

```python
#1st quantile
Q1=df.quantile(0.25)

# 3rd quantile
Q3=df.quantile(0.75)

#IQR
IQR=Q3 - Q1
dff=df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("Old DataFrame data in Rows and Column:",df.shape)
print("\nNew DataFrame data in Rows and Column:",dff.shape)
print("\nTotal Dropped rows:",df.shape[0]-dff.shape[0])
```

```python
#Percentage Data Loss using IQR
data_loss1 = (30-21)/30*100
data_loss1
```

#### WE CHECK MULTICOLLINEARITY BY VARIANCE INFLATION FACTOR (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_cal():
    vif=pd.DataFrame()
    vif["VIF FACTOR"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    vif["FEATURES"]=x.columns
    print (vif)
```

```python
vif_cal()
```

<h2 style='text-align:center;font-size:30px;background-color:Black;border:30px;color:white'> *Now Lets Train and Test the model using Regression Models* <h2>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

# LINEAR REGRESSION

```python
lr=LinearRegression()
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    print (f"At random state {i},the training accuracy is :- {r2_score(y_train,pred_train)}")
    print (f"At random state {i},the testing accuracy is :- {r2_score(y_test,pred_test)}")
    print ("\n\n")
```

#### I am Going to take random state as 6 bcz it is giving the best score .. for training it is giving =93% and for testing=94% 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=6)
```

```python
lr.fit(x_train,y_train)
```

```python
pred_test=lr.predict(x_test)
```

```python
print (r2_score(y_test,pred_test))

# r2_score is 94%
```

```python
print (' **ERROR** ')

print ('Mean Absolute Error :->',mean_absolute_error(y_test,pred_test))
print ('Mean Squared Error :->',mean_squared_error(y_test,pred_test))

print ('Root Mean Squared Error :->',np.sqrt(mean_squared_error(y_test,pred_test)))
```

## Lets do CROSS VALIDATION of Linear Regression model

```python
Train_accuracy=r2_score(y_train,pred_train)
Test_accuracy=r2_score(y_test,pred_test)

from sklearn.model_selection import cross_val_score

for i in range(2,16):
    cv_score=cross_val_score(lr,x,y,cv=i)
    cv_mean=cv_score.mean()
    print (f"At cross fold {i} the cv score is {cv_mean} and accuracy score for training is {Train_accuracy} and accuracy for testing is {Test_accuracy}")
    print ("\n\n")
```
